```python
import matplotlib.pyplot as plt
import numpy as np

plt.style.use('_mpl-gallery-nogrid')

# make a stream function:
X, Y = np.meshgrid(np.linspace(-3, 3, 256), np.linspace(-3, 3, 256))
Z = (1 - X/2 + X**5 + Y**3) * np.exp(-X**2 - Y**2)
# make U and V out of the streamfunction:
V = np.diff(Z[1:, :], axis=1)
U = -np.diff(Z[:, 1:], axis=0)

# plot:
fig, ax = plt.subplots()

ax.streamplot(X[1:, 1:], Y[1:, 1:], U, V)

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# List all available styles
print("Available styles:", plt.style.available)

# Create sample data
x = np.linspace(0, 10, 100)
y = np.sin(x)

# Create a figure with subplots to show different styles
fig, axes = plt.subplots(3, 2, figsize=(12, 10))
axes = axes.flatten()

# Display plots with different styles
styles = ['default', 'ggplot', 'seaborn-v0_8', 'bmh', 'dark_background', 'fivethirtyeight']

for ax, style in zip(axes, styles):
    with plt.style.context(style):
        ax.plot(x, y)
        ax.set_title(f"Style: {style}")
        ax.set_xlabel("x")
        ax.set_ylabel("sin(x)")
        ax.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from numpy.ma import masked_where

# Create grid data
x = np.linspace(-3.0, 3.0, 100)
y = np.linspace(-3.0, 3.0, 100)
X, Y = np.meshgrid(x, y)

# Example function: Z = sinc(r)
R = np.sqrt(X**2 + Y**2)
Z = np.sin(R) / (R + 1e-8)  # Avoid division by zero

# Mask out a circular region where R < 1
Z_masked = masked_where(R < 1, Z)

# Create contour plot
fig, ax = plt.subplots(figsize=(6, 5))
contour = ax.contourf(X, Y, Z_masked, levels=20, cmap='viridis')

# Add colorbar
cbar = plt.colorbar(contour, ax=ax)
cbar.set_label('Function Value')

# Add masked region outline for clarity
circle = plt.Circle((0, 0), 1, color='red', fill=False, linewidth=2)
ax.add_artist(circle)

ax.set_title("Contour Plot with Masked Region")
ax.set_xlabel("X-axis")
ax.set_ylabel("Y-axis")
ax.set_aspect('equal')

plt.show()
```
